## Bonus: Vztahy a známky

Uvažuj data v souboru (student-mat.csv)[student-mat.csv]. 

Jedná se o data, která obsahují výsledky studentů matematiky na portugalských školách. 

Chceme ověřit, zda je statistické rozdělení výsledků studentů a studentek ovlivněno tím, jestli mají romatický vztah.


```python
import pandas
from scipy import stats

data = pandas.read_csv("student-mat.csv")
```

Výsledky z testu jsou uloženy ve sloupci `G3` a informace o tom, jestli má student/ka milostný vztah, je ve sloupci `romantic`. Rozděl data do dvou tabulek - v jedné tabulce budou studenti a studentky mající vztah a ve druhém ti, kteří vztah nemají. 

```python
data_yes = data[data["romantic"] == "yes"]
data_no = data[data["romantic"] == "no"]

# data_yes.to_csv("data_yes.csv")
# data_no.to_csv("data_no.csv")

print(f"Počet lidí se vztahem: {data_yes.shape[0]}, počet lidí bez vztahu: {data_no.shape[0]}")
```

Ověř, kolik je v každé tabulce záznamů. 

Pro lepší představu si můžeš vytvořit z obou tabulek graf histogramu nebo odhad distribuční funkce, obojí jsme dělali na 5. lekci. 

Naším úkolem je zjistit, jestli je rozdíl obou grafů natolik velký, abychom mohli říct, že to není náhoda.



```python
bins = range(0, 20)
data_yes["G3"].hist(bins=bins)
```

```python
data_no["G3"].hist(bins=bins)
```

```python
data_yes["G3"].plot.kde()
data_no["G3"].plot.kde()
```

Dále se drž postupu, který jsme si ukazovali. 

Jde nám po porovnání statistických rozdělení. 

Jaká bude nulová a jaká bude alternativní hypotéza? A jaký test využijeme? 


Podívej se na seznam testů nebo na rozhodovací strom a vyber vhodný test. 

Vhodný test, definice hypotéz, výsledek výpočtu a interpretace jsou [zde](napoveda.ipynb).

Je třeba uvážit kolik máme souborů dat a zda jsou data párová nebo ne.

```python
# Použijeme Mann Whitney test.

# - H0: Rozdělení výsledků studentů se vztahem a studentů bez vztahu jsou stejná
# - H1: Rozdělení výsledků studentů se vztahem a studentů bez vztahu jsou stejná
```

Po výběru vhodného testu zkus tento test provést. 

Jaká je p-hodnota? A co to znamená?

```python
stats.mannwhitneyu(data_yes["G3"], data_no["G3"])
```

```python
# p-hodnota je větší než 5 %. 
# Na hladině významnosti 5 % nezamítáme H0, tj. neprokázali jsme, že by výsledky těch dvou skupin měly různé rozdělení.
```
